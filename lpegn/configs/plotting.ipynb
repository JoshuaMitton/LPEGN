```python
df = pd.read_csv(summary_dir+"/exp_summary.csv")
```

```python
print(list(df.columns.values[-11:]))
```

```python
train_accuracy = np.array(df["train_accuracy"])
test_accuracy = np.array(df["test_accuracy"])
train_loss = np.array(df["train_loss"])
test_loss = np.array(df["test_loss"])
```

```python
print(len(train_accuracy))
print(len(test_accuracy))
print(len(train_loss))
print(len(test_loss))
```

```python
train_acc = train_accuracy.reshape(10,201)
train_acc = train_acc[:,:200]
print(train_acc.shape)
test_acc = test_accuracy.reshape(10,201)
test_acc = test_acc[:,:200]
print(test_acc.shape)
train_loss = train_loss.reshape(10,201)
train_loss = train_loss[:,:200]
print(train_loss.shape)
test_loss = test_loss.reshape(10,201)
test_loss = test_loss[:,:200]
print(test_loss.shape)
```

```python
print(f'Final test accuracy : {test_acc[:,-1]}')
print(f'Final test accuracy : {np.mean(test_acc[:,-1])}')
print(f'Final test accuracy : {np.std(test_acc[:,-1])}')
```
